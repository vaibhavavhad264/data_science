```python
import pandas as pd
from numpy.conftest import dtype

df = pd.read_csv("weather_by_cities.csv")
df
```

```python
g = df.groupby('city')
print(g)
```

```python
for city, data in g:
    print(city, ":", data.temperature.max())
```

```python
def grouper(df, idx, cols):
    if 80 <= df[cols].loc[idx] <= 90:
        return "80 - 90"
    elif 50 <= df[cols].loc[idx] <= 60:
        return "50 - 60"
    else:
        return "other"
```

```python
g = df.groupby(lambda x : grouper(df, x, 'temperature'))
g
```

```python
for key, d in g:
    print(key)
    print(d)
```
