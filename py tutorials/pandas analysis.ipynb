```python
import pandas as pd
df_movies = pd.read_excel("movies_db.xlsx", "movies") 
df_movies.head(4)
```

```python
df_financials = pd.read_excel("movies_db.xlsx", "financials", converters = { 'currency' : corrected_currency}) 
df_financials.head(4)
```

```python
def corrected_currency(curr):
    if curr == "INR" or curr == "$$" or curr == "Dollars":
        return "USD"
    else:
        return "USD"

df_financials["currency"] = df_financials["currency"].apply(lambda x: corrected_currency(x))
```

```python
df_financials.head()
```

```python
df_financials["currrency"] = df_financials.drop(columns = "currrency", inplace = True)
```

```python
df_financials.head()
```

```python
del df_financials["currrency"]
```

```python
df_financials.head()
```

```python
df_merged = pd.merge(df_movies, df_financials, on = "movie_id")
df_merged.head()
```

```python
df_merged.to_excel("movies_merged.xlsx", sheet_name = "mearged", index = False) 
```
