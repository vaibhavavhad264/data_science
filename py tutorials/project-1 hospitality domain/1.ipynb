# Data Transformation

```python
df_fact_agg_booking.head()
```

```python
df_fact_agg_booking.isnull().sum()
```

```python
df_fact_agg_booking = df_fact_agg_booking.fillna({
    'capacity' : df_fact_agg_booking['capacity'].mean()
})
```

```python
df_fact_agg_booking.capacity.isnull().sum()
```

```python
df_fact_agg_booking.head()
```

```python
print(df_fact_agg_booking[df_fact_agg_booking.successful_bookings > df_fact_agg_booking.capacity])
```

```python
df_fact_agg_booking['occ_ptc'] = df_fact_agg_booking.successful_bookings / df_fact_agg_booking.capacity
```

```python
print(df_fact_agg_booking.head())
```

```python
df_fact_agg_booking['occ_ptc'] = df_fact_agg_booking['occ_ptc'].apply(lambda x : round(x*100, 2))
```

```python
print(df_fact_agg_booking.head())
```
