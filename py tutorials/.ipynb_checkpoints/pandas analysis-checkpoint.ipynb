```python
import pandas as pd
df_movies = pd.read_excel("movies_db.xlsx", "movies") 
df_movies.head(4)
```

```python
df_financials = pd.read_excel("movies_db.xlsx", "financials") 
df_financials.head(4)
```

```python
def corrected_currency(curr):
    if curr == "INR" or curr == "$$" or curr == "Dollars":
        return "USD"
    else:
        return "USD"

df_financials["currency"] = df_financials["currency"].apply(lambda x: corrected_currency(x))
```

```python
df_financials.head()
```

```python
df_financials["currrency"] = df_financials.drop(columns = ["currrency"])
```
