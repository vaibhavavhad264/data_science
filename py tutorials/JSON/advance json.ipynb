```python
import time

def square_list(number):
    start = time.time()
    result = []
    for i in number:
        result.append(i ** 2)
    end = time.time()
    return f"{(end - start) * 1000} ml second"
```

```python
def cube_list(number):
    start = time.time()
    result = []
    for i in number:
        result.append(i ** 3)
    end = time.time()
    return f"{(end - start) * 1000} ml second"
```

```python
lst = range(1000000)
square_list(lst)
```

```python
cube_list(lst)
```
