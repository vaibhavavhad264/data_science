```python
import pandas as pd
ind_df = pd.DataFrame({
    'city' : ["Nashik", "Pune", "Mumbai"],
    'temperature' : [39, 49, 38],
    'humidity' : [78, 67, 81]
})

us_df = pd.DataFrame({
    'city' : ["New York", "Diego", "Chicago"],
    'temperature' : [69, 76, 98],
    'humidity' : [98, 87, 61]
})
```

```python
df = pd.concat([ind_df,us_df], ignore_index = True)
df
```

```python
df = pd.concat([ind_df, us_df], keys = ['India', 'US'])
df
```

```python
temp_df = pd.DataFrame({
    'city' : ["Nashik", "Pune", "Mumbai", "Kolkata"],
    'temperature' : [39, 49, 38, 54]
}, index = [0, 1, 2, 3])
print(temp_df)
```

```python
hum_df = pd.DataFrame({
    'city' : ["Nashik", "Pune", "Mumbai"],
    'humidity' : [79, 65, 71]
}, index = [0, 1, 2])
print(hum_df)
```

```python
print(pd.concat([temp_df, hum_df], axis = 1))
```

```python
df = pd.merge(temp_df, hum_df, on = "city", how = "left")
print(df)
```
