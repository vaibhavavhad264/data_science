```python
import numpy as np

a = np.array([
    [101, "Ram"],
    [102, "Shyam"],
    [103, "Karna"]
])

b = np.array([
    [101, 150.50, "2023-08-01"],
    [102, 125.50, "2023-08-02"],
    [103, 250.50, "2023-08-03"]
])
```

```python
np.hstack((a, b))
```

```python
c = np.array([
    [104, "Amit"],
    [105, "Babu"],
    [106, "Kala"]
])
```

```python
np.vstack((a, c))
```
